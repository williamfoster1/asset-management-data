# ARK ETF Data Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
returns = pd.read_csv('returns.csv', header=0, index_col=0, parse_dates=True, na_values=-99.99)
prices = pd.read_csv('prices.csv', header=0, index_col=0, parse_dates=True, na_values=-99.99)
wealth = pd.read_csv('wealth.csv', header=0, index_col=0, parse_dates=True, na_values=-99.99)
volumes = pd.read_csv('volumes.csv', header=0, index_col=0, parse_dates=True, na_values=-99.99)
```

```python
prices.plot(figsize=(12, 6), title='Plot of ETF Prices From 2020-2024', alpha=0.4)
```

```python
wealth.plot(figsize=(12, 6), title='Plot Of $1000 Investment In Each ETF From 2020-2024', alpha=0.4)
```

```python
figure, axes = plt.subplots(2, 2, figsize=(12,6), edgecolor='cornflowerblue')
figure.suptitle('Distribution Of Daily Returns Of ARK ETFs')

sns.histplot(ax=axes[0, 0], data=returns.ARKF, bins=100, kde=True, stat='density', kde_kws=dict(cut=3), alpha=.4, edgecolor=(1, 1, 1, .4), label="ARKF")
sns.histplot(ax=axes[0, 1], data=returns.ARKK, bins=100, kde=True, stat='density', kde_kws=dict(cut=3), alpha=.4, edgecolor=(1, 1, 1, .4), label="ARKK")
sns.histplot(ax=axes[1, 0], data=returns.ARKQ, bins=100, kde=True, stat='density', kde_kws=dict(cut=3), alpha=.4, edgecolor=(1, 1, 1, .4), label="ARKQ")
sns.histplot(ax=axes[1, 1], data=returns.ARKW, bins=100, kde=True, stat='density', kde_kws=dict(cut=3), alpha=.4, edgecolor=(1, 1, 1, .4), label="ARKW")

axes[0, 0].set_facecolor('ivory')
axes[0, 1].set_facecolor('ivory')
axes[1, 0].set_facecolor('ivory')
axes[1, 1].set_facecolor('ivory')

axes[0, 0].axvline(x=returns.ARKF.mean(), ls="--", color="red")
axes[0, 1].axvline(x=returns.ARKK.mean(), ls="--", color="red")
axes[1, 0].axvline(x=returns.ARKQ.mean(), ls="--", color="red")
axes[1, 1].axvline(x=returns.ARKW.mean(), ls="--", color="red")
```

```python
ax = volumes.rolling(window=28).mean().ARKK.plot(figsize=(12, 6), linewidth=2, alpha=1, title='Rolling 28 Day Mean Trading Volume For ARK Innovation ETF From 2020-2024')
volumes.ARKK.plot(ax=ax, alpha=0.4)
ax.ticklabel_format(axis='y', style='plain')
ax.set_facecolor('whitesmoke')
plt.grid()
```

```python
#plt.style.use('default')
ax = prices.rolling(window=28).mean().ARKF.plot(figsize=(12, 6), linewidth=2, alpha=1, title='Rolling 28 Day Mean Price For ARK Innovation ETF From 2020-2024')
prices.ARKF.plot(ax=ax, alpha=0.4)
ax.set_facecolor('whitesmoke')
plt.grid()
```
